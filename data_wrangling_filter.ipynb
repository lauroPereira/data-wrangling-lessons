#imports

```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
df = sns.load_dataset('tips')
```

#Filtering

##Filter convencional

```python
df[df.tip > 8].head()
```

##Filter com query

```python
#A vantagem é o uso de variáveis
min_aceitavel = 8
df.query('tip > @min_aceitavel').head()
```

## Filtro convencional com multiplos filtros

```python
df[(df['tip'] > 8) & (df['smoker'] == 'Yes')].head()
```

##Filtro com query e multiplos filtros

```python
min_aceitavel = 8
is_smoker = 'Yes'

df.query('tip > @min_aceitavel & smoker == @is_smoker').head()
```
